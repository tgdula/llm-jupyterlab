```python
%%ai chatgpt -f code
Please initialise a sample pandas dataframe with a multiindex, containing daily prices (open, high, low and close) of a few stocks, over a few days. Please write a code to filter the dataframe: first for given date range, then for given date range and days
```

```python
import pandas as pd
import numpy as np

dates = pd.date_range(start='2022-01-01', periods=5)
stocks = ['AAPL', 'GOOGL', 'AMZN']
data = np.random.randn(5, 12)
midx = pd.MultiIndex.from_product([dates, stocks], names=['Date', 'Stock'])
df = pd.DataFrame(data, index=midx, columns=['Open', 'High', 'Low', 'Close'] * 3)

# Filter for given date range
filtered_df_date_range = df.loc['2022-01-02':'2022-01-04']

# Filter for given date range and days
filtered_df_date_range_days = df.loc[(['2022-01-02', '2022-01-04'], ['AAPL', 'GOOGL']), :]

filtered_df_date_range
filtered_df_date_range_days
```
